# Creating an image with the cv

```python
import cv2 
import numpy as np

## Creating blank image with numpy array

image = np.zeros((800,1000,3))

## open cv data type
image = image.astype(np.uint8)

##first row
image[0:200,0:200] = [255,255,255]
image[200:400,200:400]= [255,255,255]
image[0:200,400:600]= [255,255,255]

## second row
image[200:400,600:800]= [255,255,255]
image[0:200,800:1000]= [255,255,255]

## Third row
image[400:600,0:200]= [255,255,255]
image[400:600,400:600]= [255,255,255]
image[400:600,800:1000]= [255,255,255]

#last row
image[600:,200:400]= [255,255,255]
image[600:,600:800]= [255,255,255]

## showing image
cv2.imshow("image",image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Saving image
cv2.imwrite("Created_img.png",image)
```
